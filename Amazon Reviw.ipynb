```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from textblob import TextBlob # type: ignore
from sklearn.ensemble import RandomForestClassifier

# Load dataset
df = pd.read_csv(r'C:\Users\welcome\OneDrive\Desktop\Cleaned_amazon_reviews.csv')

# Convert reviewTime to datetime format
df['reviewTime'] = pd.to_datetime(df['reviewTime'])

# 1. Correlation Analysis
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# 2. Trend Analysis Over Time
plt.figure(figsize=(12, 6))
df.groupby(df['reviewTime'].dt.to_period('M')).agg({'overall': 'mean'}).plot(kind='line')
plt.title('Trend of Average Ratings Over Time')
plt.xlabel('Date')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()

# 3. Clustering Analysis
features = df[['overall', 'day_diff', 'helpful_yes', 'helpful_no', 'total_vote']]
features = features.fillna(0)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
kmeans = KMeans(n_clusters=3, random_state=0)
df['cluster'] = kmeans.fit_predict(scaled_features)
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='day_diff', y='overall', hue='cluster', palette='viridis')
plt.title('Clusters of Reviews')
plt.xlabel('Days Since Review')
plt.ylabel('Overall Rating')
plt.grid(True)
plt.show()

# 4. Sentiment Analysis
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity
df['sentiment'] = df['reviewText'].apply(lambda x: get_sentiment(str(x)))
plt.figure(figsize=(10, 6))
sns.histplot(df['sentiment'], bins=20, kde=True)
plt.title('Sentiment Distribution of Reviews')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 5. Feature Importance Analysis
X = df[['day_diff', 'helpful_yes', 'helpful_no', 'total_vote']]
y = df['overall']  # Assuming 'overall' is the target
model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X, y)
importances = model.feature_importances_
plt.figure(figsize=(10, 6))
sns.barplot(x=X.columns, y=importances)
plt.title('Feature Importances')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import StandardScaler

# Load the dataset
df = pd.read_csv(r'C:\Users\welcome\OneDrive\Desktop\amazon_reviews.csv')

# Display the first few rows of the dataset
print("Initial Data:")
print(df.head())

# Feature Engineering

# 1. Extracting features from the review text
# Convert review text to lowercase
df['reviewText'] = df['reviewText'].str.lower()

# Create new features: length of the review and number of words
df['review_length'] = df['reviewText'].apply(len)
df['review_word_count'] = df['reviewText'].apply(lambda x: len(x.split()))

# 2. Handling Date Features
# Convert reviewTime to datetime format
df['reviewTime'] = pd.to_datetime(df['reviewTime'], errors='coerce')

# Extract year, month, and day from reviewTime
df['review_year'] = df['reviewTime'].dt.year
df['review_month'] = df['reviewTime'].dt.month
df['review_day'] = df['reviewTime'].dt.day

# 3. Encoding Categorical Features
# Convert categorical features to numerical using one-hot encoding
df_encoded = pd.get_dummies(df[['reviewerName']], drop_first=True)

# 4. Creating Interaction Features
# Example: Interaction between review length and word count
df['interaction_length_word_count'] = df['review_length'] * df['review_word_count']

# 5. Normalizing Numerical Features
# Normalize numerical features using StandardScaler
scaler = StandardScaler()
df[['review_length', 'review_word_count', 'interaction_length_word_count']] = scaler.fit_transform(
    df[['review_length', 'review_word_count', 'interaction_length_word_count']]
)

# Combine the encoded features with the original dataframe
df = pd.concat([df, df_encoded], axis=1)

# Drop columns that are no longer needed
df.drop(['Unnamed: 0', 'reviewText', 'reviewTime'], axis=1, inplace=True)

# Display the transformed dataset
print("\nFeature Engineered Data:")
print(df.head())

# Save the feature engineered data to a new CSV file
df.to_csv(r'C:\Users\welcome\OneDrive\Desktop\Feature_Engineered_amazon_reviews.csv', index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Load the dataset
df = pd.read_csv(r'C:\Users\welcome\OneDrive\Desktop\amazon_reviews.csv')

# Display the first few rows of the dataset
print("Initial Data:")
print(df.head())

# Identify categorical columns
categorical_columns = ['reviewerName', 'reviewText', 'reviewTime']  # Example columns

# One-Hot Encoding
# Use ColumnTransformer to apply OneHotEncoder to specific columns
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(sparse_output=False, drop='first'), categorical_columns)
    ],
    remainder='passthrough'  # Keep the remaining columns unchanged
)

# Fit and transform the data
df_encoded = preprocessor.fit_transform(df)

# Extract feature names from the transformer
column_names = preprocessor.get_feature_names_out()

# Ensure the length of column names matches the number of columns in df_encoded
if len(column_names) != df_encoded.shape[1]:
    raise ValueError(f"Column names length ({len(column_names)}) does not match encoded data columns ({df_encoded.shape[1]})")

# Create DataFrame with the encoded data and correct column names
df_encoded = pd.DataFrame(df_encoded, columns=column_names)

# Display data after one-hot encoding
print("\nData After One-Hot Encoding:")
print(df_encoded.head())

# Save the encoded data to a new CSV file
df_encoded.to_csv(r'C:\Users\welcome\OneDrive\Desktop\Encoded_amazon_reviews.csv', index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Load the dataset
df = pd.read_csv(r'C:\Users\welcome\OneDrive\Desktop\amazon_reviews.csv')

# Display the first few rows of the dataset
print("Initial Data:")
print(df.head())

# Identify features and target variable
# Assuming 'overall' is the target variable for classification
features = df.drop('overall', axis=1)  # Drop the target variable from features
target = df['overall']  # Target variable

# Identify categorical columns
categorical_columns = ['reviewerName', 'reviewText', 'reviewTime']  # Example columns

# Preprocess the data
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(sparse_output=False, drop='first'), categorical_columns)
    ],
    remainder='passthrough'  # Keep the remaining columns unchanged
)

# Fit and transform the data
features_encoded = preprocessor.fit_transform(features)

# Convert the encoded data to DataFrame
feature_names = preprocessor.get_feature_names_out()
features_encoded_df = pd.DataFrame(features_encoded, columns=feature_names)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_encoded_df, target, test_size=0.3, random_state=42)

# Initialize the model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))

# Save the model if needed
import joblib
joblib.dump(model, r'C:\Users\welcome\OneDrive\Desktop\random_forest_model.joblib')

print("\nModel saved as 'random_forest_model.joblib'")
```

**Project Report: Amazon Reviews Classification**

**1. Project Overview**

The goal of this project was to build and evaluate a machine learning model to classify Amazon reviews into different rating categories (from 1 to 5) based on various features of the reviews.

**2. Data Description**

The dataset includes the following columns:
- **Unnamed: 0**: Index column (not used for modeling)
- **reviewerName**: Name of the reviewer
- **overall**: Rating given by the reviewer (target variable)
- **reviewText**: Text of the review
- **reviewTime**: Time of the review
- **day_diff**: Number of days since the review was written
- **helpful_yes**: Number of "helpful" votes received
- **helpful_no**: Number of "not helpful" votes received
- **total_vote**: Total number of votes received
- **score_pos_neg_diff**: Difference between positive and negative scores
- **score_average_rating**: Average rating score
- **wilson_lower_bound**: Wilson score interval lower bound
